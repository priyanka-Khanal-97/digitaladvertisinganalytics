All the predictor variables(Reach, Frequency, Clicks and Impressions) have a linear relationship but they are not very strong. I would not run linear regression on the predictor variable beside Clicks beacuse clicks in this model has the strongest relationship with Amount spent than the others. 

```
varr = mainFile[["Amount Spent (USD)",'Reach', 'Frequency',"Impressions","Clicks","Unique Clicks","Page Likes","Page Engagement","Post Engagement","Post Likes","Post Comments","Post Shares","Website Clicks"]].copy()
varr.head()
dummies = pd.get_dummies(mainFile.iloc[:,3])
regg = varr.join(dummies)
regg.head()
```

```
import statsmodels.api as sm

import statsmodels.formula.api as smf
target = regg.iloc[:,0]
dep = regg.iloc[:,1:395]
mod=sm.OLS(target,dep)
ols = mod.fit()
print(ols.summary())
```

The three predictors with the highest predictive value are:
1. Travel Pony: Conversion Test (47.5247)
2. Simple Multi_Fill_Campaign_2_559 ( 19.0025)
3. Page Likes(16.37)

The top 2 among the top 3 predictors are Campaign names. If we do not take that in consideration, the top 3 predictors are:
1. Page Likes (16.37)
2. Post Engagement (14.24)
3. Frequency (7.6)
